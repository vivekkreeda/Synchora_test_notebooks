## Loading the Dataset

```python
df = pd.read_csv('/content/UCI_Syncora_Synthetic.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

## Basic Overview & Data Integrity

```python

print("Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
df.describe().T
```

## Target Class Distribution

```python

sns.countplot(data=df, x='default.payment.next.month', palette='coolwarm')
plt.title('Target Distribution: Default Next Month')
plt.xticks([0, 1], ['No Default', 'Default'])
plt.ylabel("Count")
plt.show()
```

## Correlation Heatmap

```python

plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("Feature Correlation Matrix")
plt.show()
```

## Distribution Plots of Key Features

```python

sns.histplot(data=df, x='LIMIT_BAL', hue='default.payment.next.month', kde=True, element="step")
plt.title('Limit Balance Distribution by Default')
plt.show()

sns.histplot(data=df, x='AGE', hue='default.payment.next.month', kde=True, element="step")
plt.title('Age Distribution by Default')
plt.show()
```

## Boxplots for Pay History vs Default

```python

pay_cols = [f'PAY_{i}' for i in [0,2,3,4,5,6]]

plt.figure(figsize=(15, 6))
df_melt = pd.melt(df, id_vars='default.payment.next.month', value_vars=pay_cols)
sns.boxplot(data=df_melt, x='variable', y='value', hue='default.payment.next.month')
plt.title('Payment Delay History by Default Status')
plt.show()
```

# Training

```python
# Train High-Accuracy Model with AutoGluon
from autogluon.tabular import TabularPredictor

predictor = TabularPredictor(label='default.payment.next.month', eval_metric='accuracy')
predictor.fit(df, time_limit=600)  # train for max 10 minutes
```

## Model Eval

```python
# Title: Evaluate Model Accuracy & Metrics
performance = predictor.evaluate(df)
```

```python
# Title: Visualize Top Feature Importances
importance_df = predictor.feature_importance(df)
top_feats = importance_df.head(10).sort_values("importance")

plt.figure(figsize=(10,6))
sns.barplot(x=top_feats["importance"], y=top_feats.index, palette="viridis")
plt.title("Top 10 Important Features")
plt.xlabel("Importance Score")
plt.show()
```
